```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.linspace(1, 10, 100)
labels = [
    'Constant',
    'Logarithmic',
    'Linear',
    'Log Linear',
    'Quadratic',
    'Cubic',
    'Exponetial'
]
big_o = [np.ones(n.shape), np.log(n), n, n * np.log(n), n**2, n**3, 2**n]

plt.figure(figsize=(10,8))
plt.ylim(0,100)
for indice in range(len(big_o)):
    plt.plot(n, big_o[indice], label=labels[indice])

plt.legend()
plt.ylabel('Runtime')
plt.xlabel('n')
```
